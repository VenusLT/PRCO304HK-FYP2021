```python
import torch
import torch.nn as nn
from torch.autograd import Variable
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import pandas as pd

torch.manual_seed(1)    # reproducible
np.random.seed(1)
```

```python
df = pd.read_csv (r'C:\Users\vtleu\Desktop\Year 3\.ipynb_checkpoints\User1-dataset-m2.csv',header=None).T
print (df)
```

```python
# Hyper Parameters
BATCH_SIZE = 64
LR_G = 0.0001           # learning rate for generator
LR_D = 0.0001           # learning rate for discriminator
N_HOUR = 24             # think of this as number of ideas for generating an art work (Generator)
G_TEMP = 25     # it could be total point G can draw in the canvas
S_POINTS = np.vstack([np.linspace(-1, 1, G_TEMP) for _ in range(BATCH_SIZE)])
```

```python
def time_pattern():     # painting from the famous artist (real target)
    a = np.random.uniform(1, 2, size=BATCH_SIZE)[:, np.newaxis]
    pattern = a * np.power(S_POINTS, 2) + (a-1)
    pattern = torch.from_numpy(pattern).float()
    return Variable(pattern)

G = nn.Sequential(                      # Generator
    nn.Linear(N_HOUR, 500),            # random ideas (could from normal distribution)
    nn.ReLU(),
    nn.Linear(500, G_TEMP),     # making a painting from these random ideas
)

D = nn.Sequential(                      # Discriminator
    nn.Linear(G_TEMP, 500),     # receive art work either from the famous artist or a newbie like G
    nn.ReLU(),
    nn.Linear(500, 1),
    nn.Sigmoid(),                       # tell the probability that the art work is made by artist
)
```

```python
opt_D = torch.optim.Adam(D.parameters(), lr=LR_D)
opt_G = torch.optim.Adam(G.parameters(), lr=LR_G)
```

```python
for step in range(10000):
    
    pattern_sample = time_pattern()           # real painting from artist
    G_time = torch.randn(BATCH_SIZE, N_HOUR, requires_grad=True)  # random ideas
    G_pattern = G(G_time)                    # fake painting from G (random ideas)

    prob_temp1 = D(G_pattern)               # D try to reduce this prob
    
    G_loss = torch.mean(torch.log(1. - prob_temp1))  
    opt_G.zero_grad()
    G_loss.backward()
    opt_G.step()
     
    prob_temp0 = D(pattern_sample)          # D try to increase this prob
    prob_temp1 = D(G_pattern.detach())  # D try to reduce this prob
    D_loss = - torch.mean(torch.log(prob_temp0) + torch.log(1. - prob_temp1))      

    opt_D.zero_grad()
    D_loss.backward(retain_graph=True)      # reusing computational graph
    opt_D.step()

    if step % 1000 == 0:  # plotting
        plt.cla()
        plt.plot(S_POINTS[0], G_pattern.data.numpy()[0], c='#4AD631', lw=3, label='Generated hours',)
        plt.plot(S_POINTS[0], 2 * np.power(S_POINTS[0], 2) + 1, c='#74BCFF', lw=3, label='upper bound')
        plt.plot(S_POINTS[0], 1 * np.power(S_POINTS[0], 2) + 0, c='#FF9359', lw=3, label='lower bound')
        plt.text(-.5, 2.3, 'D accuracy=%.2f (0.5 for D to converge)' % prob_temp0.data.numpy().mean(), fontdict={'size': 15})
        plt.text(-.5, 2, 'D score= %.2f (-1.38 for G to converge)' % -D_loss.data.numpy(), fontdict={'size': 15})
        plt.ylim((0, 3));plt.legend(loc='upper right', fontsize=12);plt.draw();plt.pause(0.01)
        plt.show()
```
